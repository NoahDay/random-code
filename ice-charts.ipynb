```python
import geopandas as gpd
import matplotlib.pyplot as plt
import cmocean.cm as cmo
import os
from datetime import datetime


def plot_ice_chart(shapefile_path, var):
    gdf = gpd.read_file(shapefile_path)
    # Set the CRS to polar stereographic if not already set
    if gdf.crs is None or gdf.crs.to_string() != 'EPSG:3031':
        gdf = gdf.to_crs(epsg=3031)
    # Plot the data
    fig, ax = plt.subplots(figsize=(10, 10))
    gdf.plot(column=var, ax=ax, edgecolor='k', cmap=cmo.ice, legend=True)

    # Convert the string to a datetime object
    date_obj = datetime.strptime("/Users/noahday/Documents/Ice-charts/south/{}/{}.shp".format(folder, folder)[-12:-4], "%Y%m%d")

    # Customize the plot (optional)
    ax.set_title('USNIC Antarctic Sea Ice Chart ' + var + ' ' + str(date_obj))
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')
    plt.savefig(shapefile_path + '_' + var + '.png', dpi = 300, transparent=False) 
    plt.close()

directory = '/Users/noahday/Documents/Ice-charts/south'
folders = [f for f in os.listdir(directory) if os.path.isdir(os.path.join(directory, f))]
for folder in folders:
    for var in ['CT', 'CA', 'CB', 'CC', 'SO', 'SA', 'SB', 'SC', 'SD', 'FA', 'FB', 'FC', 'FS']:
        plot_ice_chart("/Users/noahday/Documents/Ice-charts/south/{}/{}.shp".format(folder, folder), var)
```
